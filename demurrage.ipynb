```python
import numpy as np
import pandas as pd
```

```python
D = 0.99
k = 1
```

```python
def demurrage(preValue):
    value = preValue * D**(k*np.log(preValue))
    return value
```

```python
A = {}
for i in range (0, 100, 10):
    if(i == 0):
        i = 1;
    temp = i
    A[i] = {}
    for j in range(0, 100, 10):
        if(j == 0):
            j = 1;
        temp = demurrage(temp)
        A[i][j] = temp
```

```python
table = pd.DataFrame(A).transpose()
```
